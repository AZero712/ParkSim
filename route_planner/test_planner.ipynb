```python
%load_ext autoreload
%autoreload 2

from dlp.dataset import Dataset
from dlp.visualizer import Visualizer

from route_planner.graph import WaypointsGraph

from pathlib import Path

import matplotlib.pyplot as plt

from route_planner.a_star import AStarPlanner
```

## Load Dataset

```python
# Load dataset
ds = Dataset()

home_path = str(Path.home())
ds.load(home_path + '/dlp-dataset/data/DJI_0012')
```

## Pyplot Visualizer

```python
# Instantiate the visualizer object
vis = Visualizer(ds)

# Plot frame
scene = ds.get('scene', ds.list_scenes()[0])
ax = vis.plot_frame(scene['first_frame'])
plt.show()
```

## Waypoints as a graph

```python
graph = WaypointsGraph()
for _, coords in vis.waypoints.items():
    graph.add_waypoint_list(coords)

# Connect sections
graph.connect(vis.waypoints['C1'][0], vis.waypoints['R1L'][-1])
graph.connect(vis.waypoints['C1'][1], vis.waypoints['R2L'][-1])
graph.connect(vis.waypoints['C1'][2], vis.waypoints['R3L'][-1])
graph.connect(vis.waypoints['C1'][3], vis.waypoints['R4L'][-1])

graph.connect(vis.waypoints['C2'][0], vis.waypoints['R1L'][0])
graph.connect(vis.waypoints['C2'][1], vis.waypoints['R2L'][0])
graph.connect(vis.waypoints['C2'][2], vis.waypoints['R3L'][0])
graph.connect(vis.waypoints['C2'][3], vis.waypoints['R4L'][0])

graph.connect(vis.waypoints['C2'][0], vis.waypoints['R1R'][-1])
graph.connect(vis.waypoints['C2'][1], vis.waypoints['R2R'][-1])
graph.connect(vis.waypoints['C2'][2], vis.waypoints['R3R'][-1])
graph.connect(vis.waypoints['C2'][3], vis.waypoints['R4R'][-1])

graph.connect(vis.waypoints['EXT'][-1], vis.waypoints['R1L'][-3])

plt_ops = {"color": 'lightgrey', "linewidth": 0.75, "linestyle": 'dotted'}
ax = graph.plot(plt_ops=plt_ops)
ax = vis.plot_frame(scene['first_frame'], ax=ax)
plt.show()
```

```python
plt_ops = {"color": 'lightgrey', "linewidth": 0.75, "linestyle": 'dotted'}
ax = graph.plot(plt_ops=plt_ops)

# Solve A* path
planner = AStarPlanner(graph.vertices[0], graph.vertices[70])
graph_sol = planner.solve()

plt_ops = {"color": 'red', "linewidth": 0.75, "linestyle": 'dotted'}
ax = vis.plot_frame(scene['first_frame'], ax=ax)
ax = graph_sol.plot(ax=ax, plt_ops=plt_ops)
plt.show()
```
